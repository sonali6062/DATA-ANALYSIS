Q1-Total number of apps in each type

```python
categories = {} # Initializing an empty dictionary to store category counts
for name in df['Category'].unique(): # Iterating through each unique category name
  ct=0 # Initializing a counter 'ct' for the current category
  for i in df['Category']: # Iterating through each category in the 'Category' column
    if(i==name): # Checking if the current category matches the unique category name
      ct+=1 # Incrementing the counter for the current category
    categories[name]=ct # Storing the count for the current category in the dictionary
# Counting the number of applications in each category and storing the counts in a dictionary
```

# HANDLING NULL VALUES USING IMPUTERS

```python
from sklearn.impute import SimpleImputer # Importing the SimpleImputer class from sklearn.impute
```

```python
data.head() # Displaying the first few rows of the DataFrame 'data'
```

```python
data.isnull().sum() # Counting the number of missing (null) values in each column of the DataFrame 'data'
```

```python
data # Displaying the entire DataFrame 'data'
```

```python
impute=SimpleImputer(missing_values=np.nan,strategy='mean') # Creating a SimpleImputer object to fill missing values (NaN) with the mean
```

Remove null from ratings

```python
impute.fit(data.iloc[:,2:3].values) # Fitting the imputer to the 'Rating' column (3rd column, index 2) to calculate the mean
```

```python
data['Rating'] = impute.transform(data.iloc[:,2:3].values) # Transforming the 'Rating' column by filling missing values with the calculated mean
```

```python
data.isnull().sum() # Counting the number of missing (null) values in each column after imputing 'Rating'
```

OBSERVATION: Null has been removed from review column

```python
data.dropna() # Dropping rows with any missing values from the DataFrame 'data' (Note: this is not assigned back to data, so the change is not permanent)
```

```python
data.isnull().sum() # Counting the number of missing (null) values in each column after the dropna() call (since it wasn't assigned, the counts are the same as before)
```

```python
imputer = SimpleImputer(missing_values=np.nan, strategy='most_frequent') # Creating a SimpleImputer object to fill missing values (NaN) with the most frequent value

# Use iloc to select all rows and all columns
data.iloc[:, :] = imputer.fit_transform(data.iloc[:, :]) # Fitting and transforming the entire DataFrame to fill all remaining missing values with the most frequent value in each column
```

```python
data.isnull().sum() # Counting the number of missing (null) values in each column after imputing all columns with the most frequent value
```

OBSERVATION- ALL NULL VALUES HAS BEEN HANDLED.
